```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import random
```

# Librería Matplotlib y Seaborn

```python
apples = [random.random() for _ in range(10)] # generamos una lista de 10 números aleatorios entre 0 y 10
oranges = [random.random() for _ in range(10)]
apples.sort(), oranges.sort() # ordenamos los números de forma ascendente
years = range(2000, 2010)
```

```python
df = pd.DataFrame({
    'apples': apples,
    'oranges': oranges},
    index = years
)
```

```python
sns.set_style("whitegrid")
```

```python
# Gráfico
plt.plot(years, apples, marker = 'x')
plt.plot(years, oranges, marker ='o')
plt.xlabel('Year')
plt.ylabel('Yields (tons per hectare)')
plt.title('Crop Yields in Kanto')
plt.legend(['Apples', 'Oranges'])
```

```python
sns.set_style("whitegrid")
years = range(2010, 2020)
grapes = [random.random() for _ in range(10)]
avocados = [random.random() for _ in range(10)]


data = pd.DataFrame({"years":years, "grapes":grapes, "avocados":avocados})
sns.barplot(x = "years", y="grapes", data=data)
```

```python
years = range(2000, 2006)
grapes = [random.random() for _ in range(6)]
avocados = [random.random() for _ in range(6)]

sns.set_style("whitegrid")
plt.bar(years, grapes, label = 'Grapes')
plt.xlabel('Year')
plt.ylabel('Yields (tons per hecatare)')
plt.title('Crop years in Kanto')
```

# Librería Pandas

## Histograma

```python
import plotly.express as px
```

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
titanic.head()
```

```python
fig = px.histogram(
    titanic,
    x="Age",
    title="Histogram of ages' passengers of Titanic",
    width = 600,
    height = 400
)
fig.show()
```

## Gráfico de barras

```python
fig = px.histogram(titanic, x="Embarked")
fig.show()
```

## Gráfico de cajas

```python
fig = px.box(titanic, x="Pclass", y="Age")

fig.update_layout(
    title="Distribution of ages for each passengers class category",
    xaxis_title = "Passenger class",
    yaxis_title = "Age"
)

fig.show()
```

## Gráfico de violín

```python
fig = px.violin(titanic, x="Sex", y="Fare")
fig.show()
```

## Visualización de relaciones
### Diagrama de dispersión

```python
fig = px.scatter(titanic, x="Fare", y="Age", template="ggplot2", color="Embarked")
fig.update_layout(
    title="Relationship between fare and age",
    xaxis_title = "Fare",
    yaxis_title = "Age"
)

fig.show()
```
